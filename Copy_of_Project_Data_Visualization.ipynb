```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales.csv')
df.info()
```

```python
print(df.isna().sum())
```

```python
df.head(5)
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
print(df.info())
```

```python
df.head(5)
```

```python
df['Hour'] = df['Time'].dt.hour
df['Minute'] = df['Time'].dt.minute
```

```python
df.head(5)
```

```python
city_profits = df.groupby('City')['gross income'].sum().reset_index()
city_profits = city_profits.sort_values(by = 'gross income', ascending=False)
```

```python
#Which branch/city is the most profitable?
plt.bar(city_profits['City'],city_profits['gross income'],color="Green",label = 'Profits')
plt.xlabel('City')
plt.ylabel('gorss income')
plt.title("Gross Income By Cities")
plt.legend()
plt.show()
```

```python
product_type_revenue = df.groupby('Product line')['Total'].sum().reset_index()
product_type_revenue= product_type_revenue.sort_values(by = 'Total')
```

```python
#Which Product line brings in the most money? Which one brings in the least?"
plt.barh(product_type_revenue['Product line'],product_type_revenue['Total'],color="Green",label = 'Revenue By Product type')
plt.xlabel('Total')
plt.ylabel('Product line')
plt.title("Revenue By Product Type")
plt.legend()
plt.show()
```

```python
counts = df['Payment'].value_counts()
print(counts)
```

```python
#What is the most popular Payment method (Cash, Ewallet, Credit Card)
plt.bar(counts.index,counts.values,color = "green",label ='Type of payments')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.title("Payment by different methods")
plt.legend()
plt.show()
```

```python
print(df['Customer type'].value_counts())
```

```python
print(df.groupby('Customer type')['Total'].mean())
```

```python
#Do we have more Members or Normal customers? Who spends more on average per transaction?
avg_spend=df.groupby('Customer type')['Total'].mean()
plt.bar(avg_spend.index,avg_spend.values,color='orange')
plt.title('Average Spend: Member vs Normal')
plt.ylabel('Average Transaction Value ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#What is the busiest hour of the day for sales? Should we add more staff during those hours?
print(df['Hour'].value_counts())
```

```python
print(df.groupby('Hour')['Invoice ID'].count())
```

```python
rush_hour=df.groupby('Hour')['Invoice ID'].count()
plt.bar(rush_hour.index,rush_hour.values,color='orange')
plt.title('Rush Hour By Time Period')
plt.ylabel('Rush')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

gender_product = df.groupby(['Product line', 'Gender'])['Total'].sum().unstack()
print(gender_product)

gender_product.plot(kind='bar', color=['pink', 'skyblue'], figsize=(10,6))

plt.title('Product Line Revenue by Gender')
plt.ylabel('Total Sales $')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
high_value_profile = df.groupby(['Customer type', 'Gender','City'])['Total'].mean()
high_value_profile = high_value_profile.sort_values(ascending=False)



heatmap_data = df.groupby(['City', 'Customer type'])['Total'].mean().unstack()


plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu")


plt.title('Average Spend by City & Member Type')
plt.show()
```

```python
#Does a higher Rating actually correlate with higher sales?
import matplotlib.pyplot as plt

plt.scatter(df['Rating'], df['Total'], alpha=0.5, color='purple')

plt.title('Relationship: Rating vs. Sales Amount')
plt.xlabel('Customer Rating (1-10)')
plt.ylabel('Transaction Total ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
